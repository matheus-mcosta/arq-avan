```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
names = ['grace', 'lunaris']

df = pd.DataFrame()

for name in names:
    tmp = pd.read_csv(f"./{name}/output_main_1.csv")
    tmp['host'] = name
    df = pd.concat([df, tmp])
    tmp = pd.read_csv(f"./{name}/output_main_2.csv")
    tmp['host'] = name

    df = pd.concat([df, tmp])

df = df.groupby(['name', 'host']).mean()
```

```python
df['total_cond_branches'] = df['BR_PRC'] + df['BR_MSP']


df['branch_miss_ratio'] = 100* df['BR_MSP'] / df['total_cond_branches']
```

```python
sns.catplot(data=df,
            x='name',
            y='branch_miss_ratio',
            hue='host',
            palette='viridis',
            kind='bar',
            legend=True
            )


# Customize the plot
plt.title('Brach Missing Percentage')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Branch Miss Ratio (%)')
plt.xlabel('')

plt.yticks([x / 10.0 for x in range(0, 8, 1)])


# Adjust layout to prevent label cutoff
plt.figure(dpi=300)
plt.show()
```

```python
sns.catplot(data=df,
            x='name',
            y='elapsed',
            hue='host',
            palette='viridis',
            kind='bar')


# Customize the plot
plt.title('Time Elapsed per execution')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Host', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to prevent label cutoff
plt.figure(dpi=300)
plt.show()
```

```python
sns.catplot(data=df,
            y='TOT_CYC',
            hue='host',
            palette='viridis',
            kind='bar',
            legend=True
            )


# Customize the plot
plt.title('Total Cycles')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Cycles')
plt.xlabel('')



# Adjust layout to prevent label cutoff
plt.figure(dpi=300)
plt.show()
```
