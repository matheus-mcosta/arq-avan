```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
names = ['grace', 'lunaris', 'joana']

df = pd.DataFrame()
df_inverted = pd.DataFrame()
final = pd.DataFrame()

for name in names:
    tmp = pd.read_csv(f"./{name}/output_main_1.csv")
    tmp['host'] = name
    df = pd.concat([df, tmp])
    # tmp = pd.read_csv(f"./{name}/output_main_2.csv")
    # tmp['host'] = name

    # df = pd.concat([df, tmp])

    # tmp = pd.read_csv(f"./{name}/output_inverted_1.csv")
    # tmp['host'] = name
    # df_inverted = pd.concat([df_inverted, tmp])
    # tmp = pd.read_csv(f"./{name}/output_inverted_2.csv")
    # tmp['host'] = name
    # df_inverted = pd.concat([df, tmp])



df = df.groupby(['name', 'host']).mean()
# df_inverted = df_inverted.groupby(['name', 'host']).mean()
# df_inverted
```

```python
df['total_cond_branches'] = df['BR_PRC'] + df['BR_MSP']
# df_inverted['total_cond_branches'] = df_inverted['BR_PRC'] + df_inverted['BR_MSP']


df['branch_miss_ratio'] = 100* df['BR_MSP'] / df['total_cond_branches']
# df_inverted['branch_miss_ratio'] = 100*df_inverted['BR_MSP'] / df_inverted['total_cond_branches']
```

```python
sns.catplot(data=df,
            x='name',
            y='branch_miss_ratio',
            hue='host',
            palette='viridis',
            kind='bar',
            legend=True
            )


# Customize the plot
plt.title('Brach Missing Percentage')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Branch Miss Ratio (%)')
plt.xlabel('')

plt.yticks([x / 10.0 for x in range(0, 8, 1)])


# Adjust layout to prevent label cutoff
plt.figure(dpi=300)
plt.show()
# # Show the plot
# sns.catplot(data=df_inverted,
#             x='name',
#             y='branch_miss_ratio',
#             hue='host',
#             palette='viridis',
#             kind='bar')


# # Customize the plot
# plt.title('Brach Missing Percentage')
# plt.xticks(rotation=45, ha='right')
# plt.yticks([x / 10.0 for x in range(0, 8, 1)])
# plt.ylabel('Branch Miss Ratio (%)')
# plt.xlabel('')

# # Adjust layout to prevent label cutoff
# plt.figure(dpi=300)
# plt.show()
# Show the plot
```
