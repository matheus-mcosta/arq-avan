```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
names = ['grace', 'lunaris', 'joana', 'luca']

df = pd.DataFrame()

for name in names:
    tmp = pd.read_csv(f"./{name}/output_main_1.csv")
    tmp['host'] = name
    df = pd.concat([df, tmp])
    tmp = pd.read_csv(f"./{name}/output_main_2.csv")
    tmp['host'] = name

    df = pd.concat([df, tmp])

df = df.groupby(['name', 'host']).mean()
```

```python
df['total_cond_branches'] = df['BR_PRC'] + df['BR_MSP']


df['branch_miss_ratio'] = 100* df['BR_MSP'] / df['total_cond_branches']

df
```

```python
sns.catplot(data=df,
            x='name',
            y='branch_miss_ratio',
            hue='host',
            palette='viridis',
            kind='bar',
            legend=True
            )


# Customize the plot
plt.title('Brach Missing Percentage')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Branch Miss Ratio (%)')
plt.xlabel('')

plt.yticks([x / 10.0 for x in range(0, 8, 1)])


# Adjust layout to prevent label cutoff
plt.figure(dpi=300)
plt.show()
```

```python
g = sns.catplot(data=df,
            x='name',
            y='elapsed',
            hue='host',
            palette='viridis',
            kind='bar',)


for container in g.ax.containers:
    g.ax.bar_label(container, fmt='%.2f', padding=2, rotation=90, color='black')

# Customize the plot
plt.title('Time Elapsed per execution')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, 50, 5))
plt.ylabel('Time Elapsed (s)')
plt.xlabel('')
# Adjust layout to prevent label cutoff
plt.figure(dpi=500)
plt.show()
```

```python

g = sns.catplot(data=df,
            x='name',
            y='TOT_INS',
            hue='host',
            palette='viridis',
            kind='bar',)


for container in g.ax.containers:
    g.ax.bar_label(container, padding=2, rotation=90, color='black')

# Customize the plot
plt.title('Total Instructions')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Instructions')
plt.xlabel('')
# Adjust layout to prevent label cutoff
```

```python
sns.catplot(data=df,
            x='name',
            y='TOT_CYC',
            hue='host',
            palette='viridis',
            kind='bar',
            legend=True
            )


# Customize the plot
plt.title('Total Cycles')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Cycles')
plt.xlabel('')



# Adjust layout to prevent label cutoff
plt.figure(dpi=300)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reset the index to make name and host regular columns
# df = df.reset_index()

# Create subplots for key metrics
fig, axes = plt.subplots(3,2, figsize=(15, 12))
fig.suptitle('Performance Metrics Comparison by File and Host', fontsize=14)

# Plot 1: Elapsed Time
sns.barplot(data=df, x='name', y='elapsed', hue='host', ax=axes[0,0])
axes[0,0].set_title('Elapsed Time')
axes[0,0].set_xticklabels(axes[0,0].get_xticklabels(), rotation=45, ha='right')

# Plot 2: Branch Miss Ratio
sns.barplot(data=df, x='name', y='branch_miss_ratio', hue='host', ax=axes[0,1])
axes[0,1].set_title('Branch Miss Ratio')
axes[0,1].set_xticklabels(axes[0,1].get_xticklabels(), rotation=45, ha='right')

# Plot 3: L1 Data Cache Misses
sns.barplot(data=df, x='name', y='L1_DCM', hue='host', ax=axes[1,0])
axes[1,0].set_title('L1 Data Cache Misses')
axes[1,0].set_xticklabels(axes[1,0].get_xticklabels(), rotation=45, ha='right')

sns.barplot(data=df, x='name', y='L2_DCM', hue='host', ax=axes[1,1])
axes[1,1].set_title('L2 Data Cache Misses')
axes[1,1].set_xticklabels(axes[1,0].get_xticklabels(), rotation=45, ha='right')


# Plot 4: Total Cycles
sns.barplot(data=df, x='name', y='TOT_CYC', hue='host', ax=axes[2,0])
axes[2,0].set_title('Total Cycles')
axes[2,0].set_xticklabels(axes[1,1].get_xticklabels(), rotation=45, ha='right')

sns.barplot(data=df, x='name', y='TOT_INS', hue='host', ax=axes[2,1])
axes[2,1].set_title('Total Instructions')
axes[2,1].set_xticklabels(axes[1,1].get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nSummary by host:")
print(df.groupby('host')[['elapsed', 'branch_miss_ratio']].mean())
```
